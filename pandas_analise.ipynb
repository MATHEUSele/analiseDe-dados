```python
import pandas as pd
# Criando um DataFrame a partir de um dicionário 
#dataframes sao estruturas de dados bidimensionais, semelhantes a tabelas, que permitem armazenar e manipular dados de forma eficiente.
dados = {"nome": ["ana","bruno","carlos",],
         "idade": [25, 47, 32,],
         "cidade": ["são paulo", "rio de janeiro", "belo horizonte",],
        }
df = pd.DataFrame(dados)
print(df)
```

```python
# Criando uma Series a partir de uma lista
#series sao estruturas de dados unidimensionais, semelhantes a uma coluna em uma tabela, que permitem armazenar e manipular dados de forma eficiente.
s = pd.Series([10, 20, 30, 40, 50])
print(s)
```

```python
# Lendo um arquivo CSV e criando um DataFrame
# O método read_csv() é usado para ler um arquivo CSV e criar um DataFrame a partir dos dados contidos no arquivo. O caminho do arquivo é passado como argumento para o método.
df = pd.read_csv('saude_mental_dados.csv')
df
```
